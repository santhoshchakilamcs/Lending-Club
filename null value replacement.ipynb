```python
from sklearn.ensemble import RandomForestRegressor

ann_inc = j.drop(['annual_inc'],axis=1)
dec_2y = j.drop(['delinq_2yrs'],axis=1)
er_cr_line = j.drop(['earliest_cr_line'],axis=1)
in_last_6mnts = j.drop(['inq_last_6mths'],axis=1)
open_ac = j.drop(['open_acc'],axis=1)
pub_r = j.drop(['pub_rec'],axis=1)
rev_u = j.drop(['revol_util'],axis=1)
tot_ac = j.drop(['total_acc'],axis=1)
ac_n_decline = j.drop(['acc_now_delinq'],axis=1)
dec_amnt = j.drop(['delinq_amnt'],axis=1)
pub_r_bank = j.drop(['pub_rec_bankruptcies'],axis=1)
tax = j.drop(['tax_liens'],axis=1)
emp_len = j.drop(['emp_length'],axis=1)
last_credit_year = j.drop(['last_credit_pull_d_year'],axis=1)
last_credit_month = j.drop(['last_credit_pull_d_month'],axis=1)
last_pymnt_year = j.drop(['last_pymnt_d_year'],axis=1)
last_pymnt_month = j.drop(['last_pymnt_d_month'],axis=1)

y_ann_inc = j['annual_inc']
y_dec_2y = j['delinq_2yrs']
y_er_cr_line = j['earliest_cr_line']
y_in_last_6mnts = j['inq_last_6mths']
y_open_ac = j['open_acc']
y_pub_r = j['pub_rec']
y_rev_u = j['revol_util']
y_tot_ac = j['total_acc']
y_ac_n_decline = j['acc_now_delinq']
y_dec_amnt = j['delinq_amnt']
y_pub_r_bank = j['pub_rec_bankruptcies']
y_tax = j['tax_liens']
y_emp_len = j['emp_length']
y_last_credit_year = j['last_credit_pull_d_year']
y_last_credit_month = j['last_credit_pull_d_month']
y_last_pymnt_year = j['last_pymnt_d_year']
y_last_pymnt_month = j['last_pymnt_d_month']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(ann_inc,y_ann_inc,test_size=0.3,random_state=1)
```

```python
model = RandomForestRegressor(random_state=1)
model.fit(x_train,y_train)
```

```python
pred_ann_inc = model.predict(x_test)
```
